<h4>Построение столбчатых гистограмм и распределения признаков для тех, кто ушёл (отток) и тех, кто остался (не попали в отток)</h4>

```python
# определим функцию, отрисовывающую гистограммы распределения по списку признаков
def get_hist_by_chorn(features):
    feature_count = len(features) # получаем количество признаков
    col = 2 # графики рисуем в 2 колонки
    row = int(feature_count/2+1) # количество строчек = половина признаков + 1, в случае нечетного числа
    i = 1 # графики начинаются с первого
    plt.figure(figsize=(20,20*col)) # поле для графиков 10*20 для каждого
    for feature in features: # перебираем все признаки
        plt.subplot(row,col,i) # определяем ячейку для размещения графика
        plt.title(f'{i}. Гистограмма c распределением по признаку "{feature}"') # заголовок графика
        plt.grid('True') # включим сетку
        
        get_bins = lambda x: x if x<100 else int(math.sqrt(x)) # подобие уравнивания количества корзин для гистограмм 
        
        churn0=data.query('churn==0')[feature] # отбираем данные признака для оставшихся
        sns.histplot(churn0, # рисуем гистограмму по этим данным
                     kde=True, # включаем график распределения
                     bins=get_bins(churn0.nunique()), # считаем количество корзин
                     label='Остались') # марка графика
        
        churn1=data.query('churn==1')[feature] # отбираем данные признака для ушедших
        sns.histplot(churn1,
                     kde=True,
                     bins=get_bins(churn1.nunique()),
                     color='darkorange', # меняем цвет графика
                     alpha=0.5, # прозрачность в половину
                     label='Ушли')
        
        plt.legend() # покажем легенду
        plt.xlabel('Значение признака') # подпишем оси 
        plt.ylabel('Частота появления признака')
        i +=1 # увеличиваем счетчик
    plt.show() # окончательно покажем результат
```

```python
# нарисуем графики по всем признакам, кроме последнего - 'churn'
get_hist_by_chorn(tuple(data.columns)[:-1])
```

<h3>Построение модели прогнозирования оттока клиентов</h3>

Построим модель бинарной классификации клиентов, где целевой признак — факт оттока клиента в следующем месяце

<h4>Разбиение данных на обучающую и валидационную выборку функцией train_test_split()</h4>

```python
X = data.drop('churn',axis=1) # разделяем выброку на признаки, удалив целевой признак
y = data['churn'] # выделяем целевой признак

# делим выборку на обучающую и тестирующую в отношении 80/20
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=0) 
```

<h4>Обучение модели на train-выборке:</h4>

<h5>Логистическая регрессия</h5>

```python
model_lr = LogisticRegression(solver='liblinear', random_state=0)
model_lr.fit(X_train, y_train)
predictions_lr = model_lr.predict(X_test)
```

<h5>Случайный лес</h5>

```python
model_rf = RandomForestClassifier(n_estimators = 50, random_state=0)
model_rf.fit(X_train,y_train)
predictions_rf = model_rf.predict(X_test)
```

<h4>Оценка метрики accuracy, precision и recall для обеих моделей на валидационной выборке</h4>

```python
# функция для подсчета и вывода требуемых метрик
def print_all_metrics(y_true, y_pred, alg):
    print('Метрики классификации по',alg)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
```

```python
print_all_metrics(y_test,predictions_lr,'LogisticRegression')
print_all_metrics(y_test,predictions_rf,'RandomForestClassifier')
```

<h4>Подытог</h4>По получившимся метрикам выбор делается в пользу модели RandomForestClassifier. 2 из 3 метрик показывают не сильно превосходящий, но лучший результат

<h4>Процент оттока из каждого кластера</h4>

```python
churn_chance = (data
    .groupby('cluster',as_index=False)['churn']
    .agg('sum')
    .merge((data
        .groupby('cluster',as_index=False)['churn']
        .agg('count')
        .rename(columns={'churn':'total'}))
          )
)
churn_chance['churn_percent'] = round(100* churn_chance['churn']/churn_chance['total'],2)
print(f'Вероятность оттока по кластерам')
display(churn_chance)
```

```python
plt.figure(figsize=(15,8))
sns.barplot(data=churn_chance,x='cluster',y='churn_percent',palette=bar_colors)
for index, cluster, val in churn_chance.drop(['churn','total'],axis=1).itertuples():
    plt.text(cluster-0.08, val+1, f'{val}%',fontsize=12)
    plt.grid(True)
    plt.title('Процент оттока из каждого кластера')
    plt.xlabel('Кластер')
    plt.ylabel('Процент оттока')
plt.show()
```

Процент оттока по кластерам лишь подтверждает выводы по распределению кластеров: более 50% представителей кластера 3 уходят в отток, и меньше 2% кластера 2 перестают посещать занятия.
