<h3><span style="color:#455a64;">Заменим названия столбцов (приведем к нижнему регистру)</span></h3>

```python
# приведем названия колонок к нижнему регистру

#for i in data.columns:
#    data.rename(columns={i:i.lower()}, inplace=True)
```

```python
data.columns = data.columns.str.lower()
```

<h3><span style="color:#455a64;">Преобразуем данные в нужные типы и обработаем пропуски</span></h3>

```python
print('Количество явных дубликатов:',data.duplicated().sum())
```

```python
# дубликаты по колонкам 'name','platform','year_of_release'

print('Неявные дубликаты по ключевым колонкам:',data[['name','platform','year_of_release']].duplicated().sum())
```

```python
# посмотрим на них

data[data[['name','platform','year_of_release']].duplicated()]
```

```python
# более подробно

data.query('name.isna() or name == "Madden NFL 13"')
```

```python
# удаляем менее информативные строки

data.drop([14244,16230],inplace=True)
```

```python
# перепроверим дубликаты по колонкам 'name','platform','year_of_release'

print('Неявные дубликаты по ключевым колонкам:',data[['name','platform','year_of_release']].duplicated().sum())
```

```python
# пропуски в колонке "год выпуска" заменим на 0 и приведем тип к int

data['year_of_release'] = data['year_of_release'].fillna(0).astype('int')
```

<p><span style="color:#455a64;">Пропуски в колонке с годом выпуска меняем на 0, чтобы при необходимости можно было их отфильтровывать, одновременно меняем тип данных на int, т.к. год записан целым числом, тем самым немного снижаем поребление памяти</span></p>

```python
# уникальные значения столбца с платформами

sorted(data['platform'].unique())
```

```python
# посмотрим возможные рейтинги

data['rating'].unique()
```

<p><span style="color:#455a64;">Проовели проверка на неявные дубликаты в категориальных столбцах.<br>Данные чистые.</span></p>

```python
# заменим пропуски на говорящее значение

data['rating'].fillna('Not Spec', inplace=True)
data['rating'].replace('K-A', 'E', inplace=True)
```

<p><span style="color:#455a64;">В колонке 'rating' меняем пропуски на значение 'Not Spec' ('Не указано'), а рейтинг 'K-A' меняем на 'E' в соотвествии с <a target="_blank" rel="noopener noreferrer" href="https://www.esrb.org/history/#:~:text=K%2DA%20(Kids%20to%20Adults)%20rating%20category%20changed%20to%20E%20(Everyone)">положением</a></span></p>

```python
# исследуем столбец user_score

data['user_score'].unique()
```

```python
print('Nan значений:', len(data.query('user_score.isna()')))
print('TBD значений:', len(data.query('user_score =="tbd"')))
print('Min значение:', data.query('~user_score.isna()')['user_score'].min())
print('Max значение:', data.query('~user_score.isna() and user_score!="tbd"')['user_score'].max())
```

```python
# заменим "tbd" значения на -1, чтобы можно было сменить тип и использовать столбец для обработки, как числовой

data['user_score'] = data['user_score'].replace('tbd','-1').astype('float64')
```

<p><span style="color:#455a64;">Проанализировали столбец 'user_score': минимальные и максимальные значения не выходят за диапазон, указанный в описании данных<br>Значение 'tbd' (To Be Determined), говорящее о том, что рейтинг неопределен, заменили на '-1'. Это поможет отделить их, если понадобится в дальнейшем. Причины же этого значения может быть в низком количестве оценок для вычисления рейтинга.<br>Замена 'tbd' позволила сменить тип данных колонки на float64</span></p>

```python
# исследуем столбец critic_score

print(data['critic_score'].unique())
print('Min значение:', data.query('~critic_score.isna()')['critic_score'].min())
print('Max значение:', data.query('~critic_score.isna()')['critic_score'].max())
```

<p><span style="color:#455a64;">Проанализировали столбец 'critic_score': минимальные и максимальные значения не выходят за диапазон, указанный в описании данных.<br>Тип данных нас устраивает.</span></p>

<h3><span style="color:#455a64;">Посчитаем суммарные продажи во всех регионах и запишем их в отдельный столбец</span></h3>

```python
# добавим колонку с общим количеством проданных копий
data['total'] = 0

for i in data.columns: # собираем колонки, которые имеют в названии 'sales'
    if i.find('sales')>0:
        data['total'] +=data[i] # и суммируем их в 'total'
```

<h4><span style="color:#455a64;">Выберем платформы с наибольшими суммарными продажами и построим распределение по годам.</span></h4>

```python
# выберем названия топ-платформ по суммарным продажам

MIN_TOTAL = 100 # определим минимальное количество продаж, млн.штук для попадания в ТОП

top_plat = (
    data
    .groupby('platform',as_index=False).agg({'total':'sum'})# группируем по плаформе и считаем сумму продаж
    .sort_values('total',ascending=False)# отсортируем по убыванию суммы
    .query('total>@MIN_TOTAL')['platform']# получим имена платформ, имеющих больше установленного миниимума продаж
)
```

```python
# определим функцию, которая будет "выравнивать" периоды графиков,
# заполняя нулями продажу в года, когда продаж не было, чтобы лучше оценивать визуализацию в одном масштабе

def fill_full_period(top_data, short_df):
    min_year = top_data.query('year_of_release != 0')['year_of_release'].min()# определяем минимальный
    max_year = top_data.query('year_of_release != 0')['year_of_release'].max()# и максимальный года в дф
    list_years = list(range(min_year,max_year+1))# делаем массив годов с мин до макс
    
    null_df = pd.DataFrame({'platform':[short_df['platform'].min()]*len(list_years),
                            'year_of_release':list_years,
                            'total':[0.0]*len(list_years)})# формируем "нулевой" датафрейм
    
    # выкидываем из нулевого ДФ строчки с годами, которые есть в переданном "коротком" ДФ
    null_df = null_df.drop(null_df.query('year_of_release in @short_df["year_of_release"]').index.values)
    # "сливаем" остатки "нулевого" и "короткий" ДФ. Возвращаем его.
    return null_df.merge(short_df, how='outer')
```

```python
top_data = (data # формируем ДФ 
            .query('platform in @top_plat and year_of_release!=0') # содержащий ТОПовые платформы
            .groupby(['platform','year_of_release'], as_index=False).agg({'total':'sum'}))
            # группируем по платформе и году, суммируем общие продажи
    
max_sales = round(top_data['total'].max()+10)# определяем максимальную продажу для лимита графиков по Y

for plat in top_plat:# для каждой ТОПовой платформы
    plt.figure(figsize=(20,5))# опредеяем поле графика
    plt.ylim(0,max_sales)# устанавливаем лимит на Y
    
    # делаем из "короткого" ДФ полный, с 0 в годах без продаж
    fill_data = fill_full_period(top_data, top_data.query('platform==@plat'))
    
    graph = sns.barplot(data=fill_data,# отрисовываем график по данным
                        x='year_of_release',
                        y='total',
                        color='steelblue',
                        ci=None)
    
    graph.set(xlabel='Год выпуска',# наводим красоту
              ylabel='Продано копий игры, млн.штук',
              title=f'Продажи игр для {plat}')
    
    plt.grid(axis='y')# нарисуем сетку для визуального восприятия данных
    plt.show()# показываем, что нарисовали
```

```python
i=1 # итератор для линий
df = pd.DataFrame(columns=['platform','year_of_release','total']) # создаем пустой ДФ для сборки данных
plt.figure(figsize=(20,10)) # подготоваливаем поле для отрисовки

for plat in top_plat: # для каждой из топовых платформ
     # делаем из "короткого" полный ДФ заполняя нулями года без продаж
    fill_data = fill_full_period(top_data, top_data.query('platform==@plat and year_of_release != 0'))
    fill_data.loc[fill_data['total']>0,'total']=i # если в какой-то год были продажи, то заменяем их на итератор
    fill_data.loc[fill_data['total']==0,'total']=np.NaN  # если продаж не было, ставив в ячейку NaN
    df = pd.concat([df,fill_data],axis=0) # добавлем в "сборочный" ДФ получившуюся таблицу
    i += 1 # увеличиваем итератор

graph = sns.lineplot(data=df, # рисуем линейный график на основе "сборочного" ДФ
                 x='year_of_release',
                 y='total',
                 hue='platform',
                 linewidth = 5,
                 palette='Paired')
graph.xaxis.set_major_locator(ticker.MultipleLocator(1)) # устанавливаем отображение каждого значения X на графике

graph.set(xlabel='Год выпуска', # наводим красоту
          ylabel='Платформа',
          title='Распределение продаж игр во времени')

plt.grid(axis='x') # отображаем сетку для Х-координат

handles, labels = plt.gca().get_legend_handles_labels() # настраиваем сортировку значений в легенде
order = list(range(len(labels)-1,-1,-1)) # такую же как порядок отображения на графике
plt.legend([handles[i] for i in order], [labels[i] for i in order]) 
plt.setp(graph.get_legend().get_texts(), fontsize='15') # немного увеличиваем шрифт в легенде

plt.axvline(2013, color='r')

plt.show() # показываем итог

del df # удаляем ненужный более ДФ
```

<h3><span style="color:#455a64;">Влияние на продажи внутри одной популярной платформы отзывов пользователей и критиков</span></h3>

<p><span style="color:#455a64;">Для рассмотрения выберем актуальную платформу с максимально заполненными оценками</span></p>

```python
 # собираем ДФ, сгруппированный по платформе и суммирующий количество отзывов критиков и пользователей и общее количество игр 
temp_df = a_data.groupby('platform',as_index=False).agg({'critic_score':'count','user_score':'count','name':'count'})
 # добавляем столбец со средним процентом заполненности оценками
temp_df['%total_score'] = round(100*((temp_df['critic_score']+temp_df['user_score'])/2)/temp_df['name'],2)
 # сортируем по убыванию процента
temp_df.sort_values('%total_score',ascending=False,inplace=True)
top_score_platform = temp_df.iloc[0,0] # первое значение первой записи - искомое. Запоминаем.
display(temp_df) # покажем, что получилось
del temp_df # удалим ненужный ДФ
```

<p><span style="color:#455a64;">Сильной корреляции между оценками критиков и пользователей и продажами не наблюдается.<br>Но! Продажи связаны меж собой: если игра продается хорошо в одном регионе, то продается так же хорошо и в другом.Т.е. если игра нравится, то нравится по всему миру всем.<br>И есть небольшая корреляция непосредственно между оценками критиков и пользователей. Это тоже не удивительно: критики - это своего рода пользователи, а пользователи - в какой-то степени критики =)</span></p>

```python
 # определим функцию, которая будет рисовать 2 диаграммы рассеяния: по критикам и по пользователям
def make_scatters(df,platform_name): # передаем датафрейм с данными и платформу, по которой нужно строить
    # фиксируем интересующие нас столбцы 
    cols = ['critic_score','user_score']
    i = 0 # задаем индикатор графика в subplots
    fig, axes = plt.subplots(1,2,figsize=(13,4)) # определяем поле для графиков из 2х ячеек
    for col in cols: # перебираем столбцы с оценками критиков и пользователей по очереди
        graph = sns.scatterplot(ax=axes[i],x=df[col],y=df['total']) # рисуем диаграмму в i-той ячейке
        who_score = 'критиков' # заготовка для наведения красоты
        if col == 'user_score':
            who_score = 'пользователей'
        graph.set(xlabel=f'Оценка {who_score}', # наводим красоту под каждую диаграмму
                  ylabel='Продано копий, млн. шт',
                  title=f'Диаграмма рассеяния для платформы {platform_name}'
                  )
        i = +1 # ячейку меняем
```

```python
make_scatters(top_score_platform_data,top_score_platform) # диаграмм для топ-платформы по проценту оценки(PC)
```

```python
# определим функцию для отрисовки графиков N-топ платформ и N-топ игровых жанров

def make_bars(region,group_tag,position,top_n): # передаем регион, таг для группировки, позицию для отображения графика и число N
    temp_df = (a_data # создаем ДФ на основе актуальных данных
            .groupby(group_tag,as_index=False).agg({region:'sum'}) # группируем по тагу и суммируем продажи по региону 
            .sort_values(region,ascending=False) # сортируем по убыванию суммы
            .head(top_n) # выбираем N-топ значений
          )
    temp_df['%'] = round(100*temp_df[region]/temp_df[region].sum(),2) # добавляем столбец с процентом от общих продаж по региону
    graph = sns.barplot(data=temp_df, # рисуем график
            x='%',
            y=group_tag,
            ax=axes[position] # в ячейке position subplot'a
            )
    ylabel = 'Платформа' # подготовка для красоты
    title=f'Топ-{top_n} игровых платформ в регионе {regions[region]}'
    if group_tag == 'genre':
        ylabel = 'Жанр игры'
        title=f'Топ-{top_n} продаж игр по жанрам в регионе {regions[region]}'
        
    graph.set(xlabel='Продано копий, % от общего количества продаж', # наводим красоту
                  ylabel=ylabel,
                  title=title,
                  xlim=(0,70)
             )
    
    for i,val in enumerate(temp_df['%']): # наводим еще немножко красоты
        graph.text(val+1,i+0.2,round(val,2))
    
```

```python
regions = {'na_sales':'Северная Америка', # определяем интересующие регионы продаж в виде словаря
           'eu_sales':'Европа',
           'jp_sales':'Япония',
           'other_sales':'Другие'}
TOP_N = 5 # определяем нужный ТОП выборки


for region in regions: # для каждого региона в списке
    fig, axes = plt.subplots(1,2,figsize=(15,3)) # подготавливаем 2 ячейки под графики
    plt.subplots_adjust(wspace= 0.4) # опредеяем расстояние между графиками
    
    make_bars(region,'platform',0,TOP_N) # рисуем график для платформы
    make_bars(region,'genre',1,TOP_N) # рисуем график для жанра
   
    plt.show() # отображаем результат
```

```python
rating_means = { # словарь рейтингов
    'E':'Для всех(E)', 
    'Not Spec':'Рейтинг не указан', 
    'M':'От 17 лет(M)', 
    'T':'От 13 лет(T)', 
    'E10+':'От 10 лет(E10+)', 
    'AO':'От 18 лет(AO)', 
    'EC':'От 3х лет(EC)', 
    'RP':'Рейтинг не присвоен(RP)'
}

rating_score = a_data.groupby('rating')[list(regions)].sum() # группируем по рейтингам суммы продаж по регионам
for region in regions:
    rating_score[region] = round(100*rating_score[region]/rating_score[region].sum(),2) # переводим в процент от общей продажи в регионе
    


graph = rating_score.T.reset_index().plot(kind='bar', # рисуем график по транспонированому ДФ
                                          x='index',
                                          stacked=True,
                                          figsize=(15,10)
                                         )
graph.set(xlabel='Регион', # делаем красоту
          ylabel='Процент продаж',
          title='Распредение возрастного рейтинга в сумме проданных игр'
         )
labels = [item.get_text() for item in graph.get_xticklabels()] # переименуем подписи оси Х 
labels = list(regions.values())
graph.set_xticklabels(labels)

 # добавим числовые значения на бары
posX=0
for col in list(rating_score.columns):
    posY=0    
    for i,val in enumerate(rating_score[col]):
        posY +=val
        if val > 1: # если значение меньше 1, то на графике подпись на него не уместится, поэтому выводить ее не стоит
            graph.text(-0.15+posX,int(posY-3),f'{val}%')
    posX +=1

 # поменяем легенду на более говорящую
l = plt.legend()
for i in range(len(l.get_texts())):
    label = rating_means.get(str(l.get_texts()[i])[12:-2])
    l.get_texts()[i].set_text(label)  
l.set_bbox_to_anchor(bbox=(1, 1))

plt.xticks(rotation=0) # повернем подписи Х
plt.show() # покажем все
```

<h2><span style="color:#455a64;">Проверка гипотез</span></h2>

<p><span style="color:#455a64;">В проверке гипотез будет использоваться t-тест для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам st.ttest_ind, т.к. данные будут предоставлены 2мя Serias и будет проводится проверка о равенстве их среднего. Параметр alternative указываться не будет, т.к. проверка будет проводится двустороняя</span></p>

```python
alpha = 0.05 # задаем альфу для проведения тестов
```

<h3><span style="color:#455a64;">Средние пользовательские рейтинги платформ Xbox One и PC одинаковые</span></h3>

<p><span style="color:#455a64;"><b>Гипотезы:</b></span></p>
<p><span style="color:#455a64;">H<sub>0</sub> Средние пользовательские рейтинги платформ Xbox One и PC одинаковые</span></p>
<p><span style="color:#455a64;">H<sub>1</sub> Средние пользовательские рейтинги платформ Xbox One и PC различаются</span></p>

```python
# выбираем нужные ГС
# убираем nan и значения '-1', которые использовались для замены 'tbd'
xone_score = data.query('platform == "XOne" and user_score > 0')['user_score'].dropna()
pc_score = data.query('platform == "PC" and user_score > 0')['user_score'].dropna()
```

```python
results = st.ttest_ind(xone_score, pc_score) # проводим тест

print('p-значение:', results.pvalue) # выводим p-value

if results.pvalue < alpha: # и результат в зависимости от сравнения
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

```python
print(xone_score.mean())
print(pc_score.mean())
print(round(100*abs(1-(xone_score.mean()/pc_score.mean())),2),'%')
```

<p><span style="color:#455a64;">Гипотеза о том, что средние пользовательские рейтинги Xbox One и PC одинаковые отвергается.<br>При простой проверке на средние можно увидеть, что разница более 7%</span></p>

<h3><span style="color:#455a64;">Средние пользовательские рейтинги жанров Action и Sports разные</span></h3>

<p><span style="color:#455a64;"><b>Гипотезы:</b></span></p>
<p><span style="color:#455a64;">H<sub>0</sub> Средние пользовательские рейтинги жанров Action и Sports разные</span></p>
<p><span style="color:#455a64;">H<sub>1</sub> Средние пользовательские рейтинги жанров Action и Sports одинаковые</span></p>

```python
# выбираем нужные ГС

action_score = data.query('genre == "Action" and user_score > 0')['user_score'].dropna()
sports_score = data.query('genre == "Sports" and user_score > 0')['user_score'].dropna()
```

```python
results = st.ttest_ind(action_score, sports_score)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

```python
print(action_score.mean())
print(sports_score.mean())
print(round(100*abs(1-(action_score.mean()/sports_score.mean())),2),'%')
```

<p><span style="color:#455a64;">Гипотеза о том, что средние пользовательские  рейтинги жанров Action и Sports разные не отвергается.<br>При простой проверке на средние можно увидеть, что разница менее 1,5%</span></p>
